So our idea is to connect suburbs of some big cities to these big cities
Then check whether the rest of cities are big or high. For example, Tula has no population and its a big city, so in this case we need manually fill the gaps
Also we will drop the rows that have LESS THAN 5 selling announcements

the towns that left after these modifications are very small so we can get rid of them

```python
import numpy as np

suburbs_mapping = {
    #Moscow oblast
    'Немчиновка': 'Москва', 'Коммунарка': 'Москва', 'Нахабино': 'Москва',
    'Власиха': 'Москва', 'Жуковка': 'Москва', 'Раздоры': 'Москва',
    'Путилково': 'Москва', 'Железнодорожный': 'Москва', 'Горки-10': 'Москва',
    'Щербинка': 'Москва', 'Посёлок завода Мосрентген': 'Москва', 'Газопровод': 'Москва',

    #Leningrad oblast
    'Шушары': 'Санкт-Петербург', 'Мурино': 'Санкт-Петербург', 'Кудрово': 'Санкт-Петербург',
    'Красное Село': 'Санкт-Петербург', 'Сестрорецк': 'Санкт-Петербург',
    'Репино': 'Санкт-Петербург', 'Парголово': 'Санкт-Петербург', 'Бугры': 'Санкт-Петербург',
    'Петергоф': 'Санкт-Петербург', 'Стрельна': 'Санкт-Петербург', 'Песочный': 'Санкт-Петербург',

    #Novosibirsk oblast
    'Краснообск': 'Новосибирск', 'Кольцово': 'Новосибирск'
}

df['city'] = df['city'].replace(suburbs_mapping)
```

```python
df['population'] = df['population'].replace(0.0, np.nan)

df['population'] = df.groupby('city')['population'].transform(lambda x: x.fillna(x.max()))
```

So we see that the 'engine' column needs modification. Right now its just a string and model wont get anything with it. So our plan is to convert this column to 3 columns
first of all - fuel type. And here is the problem, because regular cars have for example 1.2 or 5.0 volume of engine, and electric cars dont have it. So in this column we will put Nan for every electric car and not zeros, because other type of filling will lead to distorted distribution


```python
df['fuel_type'] = df['engine'].apply(lambda x: x.split(',')[-1].strip())
df['engine_volume'] = df['engine'].apply(lambda x: np.nan if 'электро' in x else float(x[:3]))
```

```python
#Here we check where л.с. letters and take what stands before them
#Take part on the right, check where is comma and take the last piece

df['engine_power'] = df['engine'].apply(lambda x: float(x.split('л.с.')[0].split(',')[-1].strip()))
```

```python
df = df.drop(columns=['engine'])
```

```python
print(df[['engine_volume', 'engine_power', 'fuel_type']].head(10))
```

Very important thing. What to do with Kwt, because as liters for regular cars its a 'volume' of engine for electric cars. But making other column will ruin statisrics, because there will be Nans for Kwt for every regular(non elecrtic) car. So what can we do.
```python
#we need to find outliers for some columns 1.5 IQR
```

```python
first_quartile = df['price'].describe()['25%']
third_quartile = df['price'].describe()['75%']

iqr = third_quartile - first_quartile
```

```python
condition = (df['price'] > (first_quartile - 1.5 * iqr)) & (df['price'] < (third_quartile + 1.5 * iqr))
```

```python
df = df[condition]
```

```python
first_quartile2 = df['year'].describe()['25%']
third_quartile2 = df['year'].describe()['75%']

iqr2 = third_quartile2 - first_quartile2
```

```python
condition2 = (df['year'] > (first_quartile2 - 1.5 * iqr2)) & (df['year'] < (third_quartile2 + 1.5 * iqr2))
df = df[condition2]
```

```python
first_quartile3 = df['mileage'].describe()['25%']
third_quartile3 = df['mileage'].describe()['75%']

iqr3 = third_quartile3 - first_quartile3
```

```python
condition3 = (df['mileage'] > (first_quartile3 - 1.5 * iqr3)) & (df['mileage'] < (third_quartile3 + 1.5 * iqr3))
df = df[condition3]
```
