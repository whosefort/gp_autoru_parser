```python
ninja_API = 'Uwi1SFJtGr0E7jkUsjiXwNW60cdQxcYKTx1wr0vH'
geo_API = '066646a31ef941b4703ae590b879a061'
density_API = 'c7cbb8c5f78c44f1b6813438181dfce3'
data = []
import logging
logging.basicConfig(level=logging.INFO, filename='api_logs.log', filemode='w')
logger = logging.getLogger(__name__)
for city in cities:
  logger.info(f'Start making requests for the {city}')
  vowel = {'city':city}
  logger.info(f'Making request of {city} to weather API of longitude and lattitude')
  req_geo = requests.get(f'http://api.openweathermap.org/geo/1.0/direct?q={city}&limit=1&appid={geo_API}', headers={'X-API-KEY':geo_API})
  req_geo.json()
  if len (req_geo.json()) == 0:
    logger.info(f'There is no such city {city} found in weather API')
    continue
  geo = req_geo.json()[0]
  lattitude = geo['lat']
  longitude = geo['lon']
  vowel['lattitude'] = lattitude
  vowel['longitude'] = longitude
  eng_name = geo['name']
  if 'local_names' in geo and 'en' in geo['local_names']:
    eng_name = geo['local_names']['en']
  logger.info(f'Making a request of population of {city} to ninjas API')
  req_city = requests.get(f'https://api.api-ninjas.com/v1/city?name={eng_name}', headers={'X-Api-Key': ninja_API})
  req_city.json()
  if type(req_city.json()) == list and len(req_city.json()) > 0:
    if 'population' in req_city.json()[0]:
      vowel['population'] = req_city.json()[0]['population']
  logger.info(f'Making a request of weather of {city} to weather API')
  req_wheather = requests.get(f'https://api.openweathermap.org/data/2.5/weather?lat={lattitude}&lon={longitude}&units=metric&appid={geo_API}', headers={'X-Api-Key':geo_API})
  req_wheather.json()
  vowel['temperature'] = req_wheather.json()['main']['temp']
  logger.info(f'Making a request of air quality of {city} to weather API')
  req_air = requests.get(f'http://api.openweathermap.org/data/2.5/air_pollution?lat={lattitude}&lon={longitude}&appid={geo_API}', headers={'X-API-KEY':geo_API})
  req_air.json()
  vowel['air_quality'] = req_air.json()['list'][0]['main']['aqi']
  logger.info(f'Making a request of elevation of {city} to meteo API')
  try:
    req_mestnost = requests.get(f'https://api.open-meteo.com/v1/elevation?latitude={lattitude}&longitude={longitude}', timeout=10)
    req_mestnost.json()
    if 'elevation' in req_mestnost.json():
      vowel['mestnost'] = req_mestnost.json()['elevation'][0]
  except:
    logger.info(f'The server of elevation does not respond, skip elevation request for the {city}')
  logger.info(f'Making a request of salary of {city} to head hunter')
  req_salary = requests.get(f'https://api.hh.ru/vacancies?text={city}&only_with_salary=true&per_page=20')
  req_salary.json()
  salary = []
  for elem in req_salary.json()['items']:
    wage = elem['salary']
    if wage != None and wage['currency'] == 'RUR':
      if wage['to'] and wage['from']:
        salary.append((wage['to'] + wage['from']) / 2)
      elif wage['to']:
        salary.append(wage['to'])
      elif wage['from']:
        salary.append(wage['from'])
  if len(salary) != 0:
    vowel['salary'] = sum(salary) / len(salary)
  logger.info(f'Making a request of car infrastructure of {city} to geoapify API')
  req_infrastructure = requests.get(f'https://api.geoapify.com/v2/places?categories=commercial.vehicle,service.vehicle&filter=circle:{longitude},{lattitude},15000&limit=500&apiKey={density_API}', headers={'X-API-KEY':density_API})
  req_infrastructure.json()
  vowel['infrastructure'] = len(req_infrastructure.json()['features'])
  data.append(vowel)
  logger.info(f'All possible data gathered for the {city}')
  time.sleep(0.5)
data
```
